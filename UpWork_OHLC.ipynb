```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as mticker

# from matplotlib.finance import candlestick_ohlc
import mplfinance as mpf

from matplotlib.pyplot import figure
%matplotlib inline


import warnings
warnings.filterwarnings('ignore')
```

```python
datafile = 'EOD_SQ_daily_ta_values.csv'
data = pd.read_csv(datafile, index_col = 'date')
data.index = pd.to_datetime(data.index) # Converting the dates from string to datetime format
data = data.tail(120)

data.head(10)
```

```python
plt.style.use('fivethirtyeight')

mpf.plot(data[-120:], type='candlestick', no_xgaps = True, figsize=(50,16))

plt.figure(figsize=(50,16))
##plt.xlabel('date',fontsize =14)
##plt.xticks(list(data.index))
plt.xticks(rotation=90)
plt.plot( 'pos_3_atr', data=data, color='#22a565', marker ='o', linewidth=2)
plt.plot( 'pos_2_atr', data=data, color='#ccaf11', marker ='o', linewidth=2)
plt.plot( 'pos_1_atr', data=data, color='#cc2311', marker ='o', linewidth=2)
plt.plot( 'ema20', data=data, color='#3acc11', marker ='o', linewidth=2)
plt.plot( 'sma20', data=data, color='#4311cc', marker ='o', linewidth=2)
plt.plot( 'neg_1_atr', data=data, color='#cc2311', marker ='o', linewidth=2)
plt.plot( 'neg_2_atr', data=data, color='#ccaf11', marker ='o', linewidth=2)
plt.plot( 'neg_3_atr', data=data, color='#22a565', marker ='o', linewidth=2)
plt.plot( 'bollinger_band_top', data=data, color='#1186cc', marker ='o', linewidth=2)
plt.plot( 'bollinger_band_bottom', data=data, color='#1186cc', marker ='o', linewidth=2)
plt.plot( 'squeeze_keltner_upper', data=data, color='#cc119d', marker ='o', linewidth=2)
plt.plot( 'squeeze_keltner_upper', data=data, color='#cc119d', marker ='o', linewidth=2)
plt.plot( 'squeeze_keltner_lower', data=data, color='#22a565', marker ='o', linewidth=2)

plt.scatter(list(data.index),'squeeze_boolean', data=data,c=data.squeeze_boolean, cmap="RdGy", marker='o', s=200)
plt.legend(fontsize=15)
plt.show()
```
